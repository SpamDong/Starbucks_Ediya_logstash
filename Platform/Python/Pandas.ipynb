```python
import pandas as pd

# csv파일 읽기
df = pd.read_csv('./rawdata.csv')

df
```

```python
df.columns
```

```python
starbucks = df[df.상호명.str.contains('스타벅스')]

starbucks
```

```python
ediya = df[df.상호명.str.contains('이디야|EDIYA')]

ediya
```

```python
starbucks = starbucks.rename(columns = {'상호명' : '브랜드명'})
starbucks = starbucks.reset_index(drop=True)

starbucks['브랜드명'] = '스타벅스'
```

```python
starbucks = starbucks[['브랜드명', '위도', '경도']]
```

```python
ediya = ediya.rename(columns = {'상호명' : '브랜드명'})
ediya = ediya.reset_index(drop=True)

ediya['브랜드명'] = '이디야'
```

```python
ediya = ediya[['브랜드명', '위도', '경도']]
```

```python
starbucks_ediya = pd.concat([ediya, starbucks], axis = 0) #1은 옆으로 0은 아래로

starbucks_ediya 
```

```python
# csv 파일 다운로드
starbucks_ediya.to_csv('./starbucks_ediya.csv', index=False)
```

```python
import json
import logging
from logging import StreamHandler
import logstash

host = '192.168.216.136'

test_logger = logging.getLogger('python-logstash-logger')

test_logger.setLevel(logging.INFO)
test_logger.addHandler(logstash.TCPLogstashHandler(host, 5000, version=1))
test_logger.addHandler(StreamHandler())

inout_f = open("starbucks_ediya.csv", 'r', encoding="UTF-8")
inout_lines = inout_f.readlines()

json_f = open("coffee.json", 'w', encoding="UTF-8")

for i in inout_lines:
    brand = i.split(',')[0]
    lon = i.split(',')[1]
    lat = i.split(',')[2]
    # print(brand)
    data = {
        "brand": brand,
        "coffee_geo": {"lat": lat, "lon": lon}
    }

    json_f.write(json.dumps(data, ensure_ascii=False) + "\n")
    test_logger.info('커피', extra=data)


inout_f.close()
json_f.close()
```
